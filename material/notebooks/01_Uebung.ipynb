### Aufgabe 1

Iterieren Sie über alle Zahlen $z$ zwischen 0 und 30 und erfüllen sie folgene Bedigungen:
- Wenn $z$ durch 3 teilbar: $z^2$ ausgeben. 
- Wenn $z$ durch 5 teilbar: $\frac{z}{2}$ ausgeben.

        print("z = ", z)

### Aufgabe 3

Gegeben sei ein String wie `s = "aabbccaeffbg"`.

Schreiben Sie eine Funktion, die, welche für jeden Buchstaben seine Häufigkeit ausgibt.

Nutzen Sie dazu 

- a) ein normales Dictionary (`dict`)
- b) ein Default-Dict (`from collections import defaultdict`)
- c) einen Counter (`from collections import Counter`)


---
<details>

<summary>→ Tipps ▼</summary>

- Fügen Sie zunächst mit einer Schleife jeden auftretenden Buchstaben als *key* zum dict hinzu und setzen Sie den zugehörigen *value* auf 0.
- Schreiben Sie eine weitere Schleife um die Buchstaben zu zählen, indem Sie den jeweiligen *value* "inkrementieren" (um eins erhöhen).
- Schreiben eine dritte Schleife für die Ausgabe. 
</details>


<details>

<summary>→ Musterlösung ▼</summary>
    
```python
def count(s):
    res = {}
    for c in s:
        res[c] = 0
    for c in s:
        res[c] += 1
    
    for key, value in res.items():
        print(f"{key}: {value}")
    
# Funktion aufrufen
count(s)
```
</details>

### Aufgabe 14

Lösen Sie das folgende lineare Gleichungssystem?


$$
\underbrace{
\left(\begin{matrix}
-1 &  -1 & -1 \\
-1  & 5  & 2\\
 3 & -4  & 2 \\
\end{matrix} \right)}_{\mathbf{A}}
\cdot
\mathbf{x}
=
\underbrace{
\left(\begin{matrix}
 -16\\
 50 \\
 -14
\end{matrix} \right)}_{\mathbf{y}}
$$


und führen Sie anschließend die Probe durch:

$$
\mathbf{A}\cdot \mathbf{x}- \mathbf{y} \stackrel{!}{=} \mathbf{0}.
$$



---

<details>

<summary>→ Tipps ▼</summary>

- `np.array(<Liste>)` → 1D-Array (≙ Vektor), `np.array(<Liste_von_Listen>)` → 2D-Array (≙ Matrix)
- `np.linalg.solve?`: Lösen linearer GS,
- `@`-Operator: Matrix-Multiplikation

</details>

---

<details>

<summary>→ Musterlösung ▼</summary>
    
```python
A = np.array([[-1, -1, -1],
              [-1,  5,  2],
              [ 3, -4,  2]])

y = np.array([-16,  50, -14])

x = np.linalg.solve(A, y)
print(x)

# Probe
eqn_err = A@x-y
print(eqn_err)
```
</details>
